```python
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score

def cross_validate(X, y, n_splits=5, n_estimators=10, max_depth=10, min_samples_split=10):
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)
    accuracies = []

    for train_index, test_index in kf.split(X):
        # Podział danych na foldy
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]

        # Budowa modelu
        rf = RandomForest(
            n_estimators=n_estimators,
            max_tree_depth=max_depth,
            min_tree_samples_split=min_samples_split,
            max_features='sqrt'
        )
        train_data = pd.concat([X_train, y_train], axis=1)
        rf.build_forest(train_data, y_train.name)

        # Predykcja i ocena
        predictions = [rf.predict(row[1]) for row in X_test.iterrows()]
        accuracy = accuracy_score(y_test, predictions)
        accuracies.append(accuracy)

    return sum(accuracies) / len(accuracies)

```

```python
# Dane (przykład)
df = pd.read_csv("../datasets/yeast/yeast.data", delim_whitespace=True, header=None)
df.columns = ["sequence_name", "mcg", "gvh", "alm", "mit", "erl", "pox", "vac", "nuc", "class"]
X = df.drop(["class", "sequence_name"], axis=1)
y = df["class"]

# Parametry walidacji
n_splits = 5
n_estimators = 100
max_depth = 20
min_samples_split = 10

# Uruchomienie walidacji
mean_accuracy = cross_validate(X, y, n_splits, n_estimators, max_depth, min_samples_split)
print(f"Średnia dokładność ({n_splits}-fold CV): {mean_accuracy:.4f}")
```
